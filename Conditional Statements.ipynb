```python
employee_1=["Amar" , 19, "Male" , 300000]
employee_2=["Sanwal", 19, "Male", 400000]
employees=[employee_1, employee_2 , ["Khurram", 24, "Male" , 200000], ["Saad", 23, "Male", 100000]]
```

```python
salary_gr_250k = 0
for i in employees:
    salary = i[-1]
    if salary >=250000:
        salary_gr_250k+=salary
```
